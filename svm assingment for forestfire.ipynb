```python
##Dropping the month and day columns
df.drop(["month","day"],axis=1,inplace =True)
```

```python
x = df.iloc[:,0:28]
y = df.iloc[:,28]
```

```python
#Preprocessing the data. As, there are categorical variables
from sklearn.preprocessing import LabelEncoder
LE= LabelEncoder()
for i in df.columns:
        df[i]= LE.fit_transform(df[i])
```

```python
# Splitting Train and Test
from sklearn.model_selection._split import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
########## support vector manchine for linear #########
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', C=1.0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_test  = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))
```

```python
########## svm for poly ########
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='poly', degree=16)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_test  = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))
```

```python
############## svm for rbf ###########
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf', gamma=2)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred_train = clf.predict(X_train)
y_pred_test  = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))
```
